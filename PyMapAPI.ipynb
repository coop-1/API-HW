```python
import pandas as pd
import requests
import time
import matplotlib.pyplot as plt
import numpy as np
from citipy import citipy
from weatherAPI import api_key
```

```python
URL = "http://api.openweathermap.org/data/2.5/weather?"
cityinfo = []
x = 0
y = 0
```

```python
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=5000)
lngs = np.random.uniform(low=-180.000, high=180.000, size=5000)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

print(len(cities))
```

```python
print ("Start Time : %s" % time.ctime())

while x < 750:
    z = 0
    while z != 1:
        params = {
        "q" : cities[y],
        "units" : "imperial",
        "appid" : api_key
        }
        response = requests.get(URL, params = params)
        code = response.status_code
        
        if code == 404:
            print("\n")
            print("Invalid City - Processing...")
            print("\n")
            y += 1
        elif code == 200:
            info = {}
            response = response.json()
            info["name"] = response["name"]
            info["country"] = response["sys"]["country"]
            info["lat"] = response["coord"]["lat"]
            info["lon"] = response["coord"]["lon"]
            info["temp_max"] = response["main"]["temp"]
            info["hum"] = response["main"]["humidity"]
            info["wind"] = response["wind"]["speed"]
            info["cloud"] = response["clouds"]["all"]
            cityinfo.append(info)
            name = "name"
            country = "country"
            val = x + 1
            print (f"City# {val} - {cityinfo[x][name]}, {cityinfo[x][country]}")
            x += 1
            z += 1
            y += 1



print ("End Time: %s" % time.ctime())
```

```python
cityinfo_df = pd.DataFrame(cityinfo)
cityinfo_df.to_csv("CityWeather.csv",index = False)
print(cityinfo_df.head())
```

```python
plt.scatter(cityinfo_df["lat"],cityinfo_df["temp_max"],marker = "o")
plt.title("Temperature in World Cities")
plt.ylabel("Temperature (F)")
plt.xlabel("Latitude")
plt.grid(True)
plt.savefig("Lat_v_Temp.jpeg")
```
